```python
# cols_inventory = [
#     'OrderList',
#     'Product',
#     'Branch Name', 
#     'Dept Fullname', 
#     'Group Fullname'
#     ]

cols_sales = [
    'Branch Name', 
    'Dept Fullname',
    ]

# df_inventory_filtered = df_inventory[cols_inventory]
df_sales_filtered = df_sales[cols_sales]

# df_inventory_filtered.head()
df_sales_filtered.groupby('Branch Name')

df_sales_filtered.head()

transactions_by_branch = df_sales_filtered.groupby('Branch Name').size().reset_index(name='Transaction Count')
print(transactions_by_branch)
```

```python
sales_by_dept_branch = df_sales_filtered.groupby([
    'Branch Name', 
    'Dept Fullname'
]).size().reset_index(name='Total Transactions')

# Afficher le résultat
print(sales_by_dept_branch)
```

```python
# Utiliser groupby().nlargest() pour obtenir le top N (ici, 3) pour chaque groupe
top_3_depts_by_branch = sales_by_dept_branch.groupby('Branch Name', group_keys=False).apply(
    lambda x: x.nlargest(2, 'Total Transactions')
)

# Afficher le résultat
print(top_3_depts_by_branch)
```
